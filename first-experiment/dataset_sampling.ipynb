```python
from ftplib import FTP
import os
import pandas as pd

source_path="Hysped/Integracia_po_lokalitach/vystupy/HyspedDataSpolu4.zip"
destination_path="HyspedDataSpolu4.zip"
```

```python
ftp_host=os.getenv("FTP_HOST")
ftp_user=os.getenv("FTP_USER")
ftp_pass=os.getenv("FTP_PASS")
ftp_port=int(os.getenv("FTP_PORT"))
```

```python
ftp = FTP()
ftp.connect(host=ftp_host, port=ftp_port)
ftp.login(user=ftp_user, passwd=ftp_pass)
ftp.retrlines('LIST')

with open(destination_path, "wb") as local_file:
    # Download the file from the FTP server
    ftp.retrbinary(f"RETR {source_path}", local_file.write)
print(f"File downloaded successfully to {os.path.abspath(destination_path)}")
```

```python
df=pd.read_csv("HyspedDataSpolu4.zip")
df.shape
```

```python
collums_to_drop = [col for col in df.columns if col.startswith("T") or col.startswith("VI") or col == "DATUM"]
df = df.drop(columns=collums_to_drop)
df.shape
list(df.columns)
```

```python
df.head()
unique_values = df['NAZ_LOKALI'].unique()
# Vypíše unikátne hodnoty na obrazovku
print("Unikátne názvy v stĺpci NAZ_LOKALI sú:")
for value in unique_values:
    print(value)
```

```python
df["DRUH_POVR"].value_counts()
```

```python
sample_fraction=0.05
smaller_df=df.groupby("DRUH_POVR",group_keys=False).apply(lambda x: x.sample(frac=sample_fraction))
```
